Порахувати всі унікальні значення в массиві та вивести їх разом із кількістю

```python
uniq_value = np.unique(df, return_counts=True)
# Виводимо унікальні значення
print('Унікальні значення : ', uniq_value[0])
# Виводимо їх кількість
print('Кількість значень : ', uniq_value[1])
```

Розбити масив по горизонталі на 2 рівні частини (не використовувати абсолютні числа, мають бути два массиви по 4 колонки)

```python
two_df = np.vsplit(df,2)
first_df = two_df[0]
second_df = two_df[1]
#перевіряємо що дійсно 2 масива по 75 строк та 2 колонки
print('Розмір першого датафрейма : ', first_df.shape)
print('Розмір другого датафрейма : ', second_df.shape)
```

Відсортувати обидва массиви по 1-й колонці: 1-й за збільшенням, 2-й за зменшенням

```python
# сортуємо перший массив за збільшенням
min_max = first_df[np.argsort(first_df[:,0])]
# сортуємо другий масив за зменшенням (сортуємо та здійснюємо реверс)
max_min = second_df[np.argsort(second_df[:,0])][::-1,]

print('Відфільтрований масив за зменшенням :\n', min_max, '\n')
print('Відфільтрований масив за збільшенням :\n', max_min)
```

Зібрати обидва массиви в одне ціле

```python
aft_sort_df = np.concatenate([min_max, max_min], axis=0)
print("Масив після об'єднаяння : \n", aft_sort_df)
```

Знайти найбільш часто повторюване значення в массиві

```python
#знаходимо індекс більш повторювального числа із масива uniq_value
more_repeat_ind = np.where(uniq_value[1] == uniq_value[1].max())[0]
#знаходимо по індексу значення та виводимо його
more_repeat_val = uniq_value[0][more_repeat_ind]
print(more_repeat_val)
```

Написати функцію, яка б множила всі значення в колонці, які менше середнього значения в цій колонці, на 2, і ділила інші значення на 4.

```python
# звичайна функциія для застосування vectorize
def do_matematic(mn, masiv):
    if masiv < mn:
        masiv = masiv * 2
    else:
        masiv = masiv / 4
    return masiv
```

Застосувати отриману функцію до 3-ї колонки

```python
final_column = aft_sort_df[:,2]
mn_final_column = np.mean(final_column)
vect_do_matematic = np.vectorize(do_matematic, otypes=[float])
final_columns = vect_do_matematic(mn_final_column, final_column)
print('Приклад роботи функції з vectorize :\n',final_columns)
```

Аналогічна функція без vectorize

```python
def mathematic(col):
    mn = np.mean(col)
    new_col = np.where(col < mn, col*2, col/4)
    return new_col
```

```python
non_vectorize = mathematic(aft_sort_df[:,2])
print(non_vectorize)
```

```python
#перевіряємо чи дорівнюють одна одній:
if final_columns.all() == non_vectorize.all():
    print('Congratulation 🥳')
```
