**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
# вирішемо це завдання 2а методами
# Перший метод - зробимо ранг освіти від 1 до 16
'''
Preschool = 1
1st-4th = 2
5th-6th = 3
7th-8th = 4
9th = 5
10th = 6
11th = 7
12th = 8
HS-grad = 9
Some-college = 10
Bachelors = 11
Prof-school = 12
Assoc-acdm = 13
Assoc-voc = 14
Masters = 15
Doctorate = 16
PS не впевнений що градація саме така, але нам важливо щоб наші значення (що ми порівнюємо були у проміжку від 11-16
'''
# Зробимо словник
rang = {'Preschool' : 1, '1st-4th' : 2, '5th-6th' : 3, '7th-8th' : 4, '9th' : 5, '10th' : 6, '11th' : 7, '12th' : 8,\
        'HS-grad' : 9, 'Some-college' : 10, 'Bachelors' : 11, 'Prof-school' : 12, 'Assoc-acdm' : 13, 'Assoc-voc' : 14,\
        'Masters' : 15, 'Doctorate' : 16}
revers_rang = {v: k for k, v in rang.items()}
# так як будемо приміняти 2 підхода будемо робити df у нових змінних
df1 = data
# підгтовка данних
df1 = df1.replace({'education':rang})

# перевіряємо чи всі данні ми замінили, чи не залишилось якихось градацій не пронумерованих
#print(df1.education.unique())
# перевіряємо чи немає елементів nan
#print(df1.education.isna().unique())

#відповідь на питання

if df1.query("salary == '>50K'").education.min() >= 11:
    print('Твердження вірне')
else:
    print('Ні твердження не відповідає дійсності. Маємо такі відсоткові данній по освіті (вісдоток освіти яка рахується середньою та нище):')
    output_s = df1.query("salary == '>50K'").education.value_counts(normalize=True)
    output_s = output_s[output_s.index <= 10].rename(index=revers_rang)
    print(output_s * 100)
```

```python
# реалізуємо інший підход (можно зробити через np.array, але зробимо через list щоб додатково не імпортувати бібліотеку)
# спочатку фідфілтруємо по зарплаті людей
df2 = data.query("salary == '>50K'")
# дивимось скільки значень education у нас є у відсотках
education_s = df2.education.value_counts(normalize=True) * 100
# робимо список значень що отримали
education_val = education_s.index.tolist()
# робимо спичок значень вищої освіти
high_education = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
# даємо відповідь на питання
if list(set(education_val) - set(high_education)) == []:
    print('Твердження вірне')
else:
    print('Ні твердження не відповідає дійсності. Маємо такі відсоткові данній по освіті (вісдоток освіти яка рахується середньою та нище):')
    print(education_s[list(set(education_val) - set(high_education))])
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
# Робимо 2 списка з женатим та холостими 
all_vall = data['marital-status'].unique().tolist()
married = []
not_maried = []
for i in range(0,len(all_vall)):
    if 'Married' in all_vall[i]:
        married.append(all_vall[i])
    else:
        not_maried.append(all_vall[i])
# Робимо виборку з нашого df по зарплаті та статі
all_series = data.query("salary == '>50K' and sex == 'Male'")
# заміняємо значення 
for i in range(0,len(married)):
    all_series = all_series.replace({'marital-status' : {married[i] : 'married'}})
for i in range(0, len(not_maried)):
    all_series = all_series.replace({'marital-status' : {not_maried[i] : 'not_married'}})
    
#дивимось чи все замінилося
#all_series['marital-status'].unique()

answer = all_series['marital-status'].value_counts(normalize=True) * 100
#виводимо відповідь
print(f'Білше доля заробляючих - {answer[answer == answer.max()].index[0]}\nВідсоткове відношення - {answer.max()}%')
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_worktime = data['hours-per-week'].max()
hard_work_df = data[data['hours-per-week'] == max_worktime]
people_hard = hard_work_df.shape[0]
rich_hard_man = hard_work_df[hard_work_df['salary'] == '>50K'].shape[0]
print(f'Максимальна кількість робочого часу складає : {max_worktime}\nЛюдей які працюють цю кількість часу : {people_hard}\n\
З них кто заробляє понад 50К : {rich_hard_man}, що складає {round(rich_hard_man/people_hard*100, 2)}% від всіх кто працює багато')
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
rich_df = data.query("salary == '>50K'").groupby(['native-country']).agg({'hours-per-week':'mean'})\
.rename(columns ={'hours-per-week': 'rich_mean_time_work'})

poor_df = data.query("salary == '<=50K'").groupby(['native-country']).agg({'hours-per-week':'mean'})\
.rename(columns ={'hours-per-week': 'poor_mean_time_work'})
out_df = pd.concat([rich_df, poor_df], axis=1)
# зробимо колонку з перевіркою (в залежності від країни) де перевіримо "Чи правда, що чим більше працюєш - тим більше заробляєш"
out_df['more_work_more_money'] = out_df.rich_mean_time_work > out_df.poor_mean_time_work
# при виводі є значення "?" в значеннях країни, змінемо його на not_identificate
out_df = out_df.rename(index = {'?' : 'not_identificate'})
out_df
    
```

```python
poor_country = ', '.join(out_df.rich_mean_time_work[out_df.rich_mean_time_work.isna()].index.tolist())
rich_country = ', '.join(out_df.poor_mean_time_work[out_df.poor_mean_time_work.isna()].index.tolist())
hardwork_bigrich = ', '.join(out_df.query("more_work_more_money == True").index.tolist())

if poor_country == '':
    poor_country = 'відсутні'
if rich_country == '':
    rich_country = 'відсутні'
if hardwork_bigrich == '':
    hardwork_bigrich = 'відсутні'


print(f"Висновки які можемо зробити:\nКраїни в яких заробітна плата не перевищує 50к взагалі : {poor_country}\n\
Країни в яких заробітна плата не опускається нижче 50К : {rich_country}\n\
Країни про які можна сказати, що чим більше працюєш - більше заробляєш : {hardwork_bigrich}")
```
