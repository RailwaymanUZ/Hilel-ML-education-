```python
import numpy as np
import requests
from io import StringIO
```

1 –ó–Ω–∞–π—Ç–∏ –≤ –¥–∞—Ç–∞—Å–µ—Ç—ñ —Ç–∞—Ä–≥–µ—Ç —Ç–∞ –≤–∏–¥–∞–ª–∏—Ç–∏ —Ü—é –∫–æ–ª–æ–Ω–∫—É –∑ –¥–∞—Ç–∞—Å–µ—Ç—É (–≤–∏–¥–∞–ª—è—Ç–∏ –∑–∞ —ñ–Ω–¥–µ–∫—Å–æ–º)

```python
respone = requests.get('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
df = np.genfromtxt(StringIO(respone.text), delimiter=',', dtype=None, encoding=None)
# –¥–∏–≤–∏–º–æ—Å—å –Ω–∞ —Å–∞–º dataframe
print(df)
```

```python
#–∑–∞–≤–∞–Ω—Ç–∞–∂—É—î–º–æ 4 —Å—Ç–æ–≤–ø—á–∏–∫–∞
df = np.loadtxt(StringIO(respone.text), delimiter=',', usecols=(0,1,2,3))
# —è–∫—â–æ –≤–∏–≤–æ–¥–∏—Ç—å—Å—è 2 —á–∏—Å–ª–∞ (150, 4) —Ü–µ –≤ –Ω–∞—Å —î 2D –º–∞—Å–∏–≤, –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ü–∏—Ñ—Ä –≤–∫–∞–∑—É—î –Ω–∞ "–º–∞—Å–∏–≤–Ω—ñ—Å—Ç—å"
print('–†–æ–∑–º—ñ—Ä Data Frame: ', df.shape) 
# –≤–∏–≤–æ–¥–∏–º–æ —Å–∞–º –º–∞—Å–∏–≤
print('–î–∞–Ω–Ω—ñ Data Frame: \n', df)
```

–ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ mean, median, standard deviation –¥–ª—è 1-—ó –∫–æ–ª–æ–Ω–∫–∏

```python
print('–°–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è : ', np.mean(df[:,0]))
print('–ú–µ–¥—ñ–∞–Ω–∞ : ', np.median(df[:,0]))
print('–°–µ—Ä–µ–¥–Ω—å–æ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è : ', np.std(df[:,0]))
```

–í—Å—Ç–∞–≤–∏—Ç–∏ 20 –∑–Ω–∞—á–µ–Ω—å np.nan –Ω–∞ –≤–∏–ø–∞–¥–∫–æ–≤—ñ –ø–æ–∑–∏—Ü—ñ—ó –≤ –º–∞—Å–∏–≤—ñ (–ø—Ä–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—ñ –∑–≤–∏—á–∞–π–Ω–æ–≥–æ —Ä–∞–Ω–¥–æ–º—É –º–æ–∂—É—Ç—å –Ω–∞–∫–ª–∞—Å—Ç–∏—Å—å –ø–æ–∑–∏—Ü—ñ—ó, —Ç–æ–º—É –∑–Ω–∞–π—Ç–∏ —Ä—ñ—à–µ–Ω–Ω—è, —è–∫–µ –≥–∞—Ä–∞–Ω—Ç—É—î 20 —É–Ω—ñ–∫–∞–ª—å–Ω–∏—Ö –ø–æ–∑–∏—Ü—ñ–π)

```python
#—Å—Ç–≤–æ—Ä–∏–º–æ Dataframe –∑ –ø–æ–∑–∏—Ü—ñ—è–º–∏ –¥–ª—è –∑–∞–º—ñ–Ω–∏ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –¥–ª—è –≤–∏–±–æ—Ä–∞ —Å—Ç—Ä–æ–∫ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º choice –∑ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–º replace=False
row = np.random.choice(range(0,150), size=20, replace=False) 
column = np.random.randint(0,3,size=20)
repl_df = np.column_stack((row,column))
print(repl_df)
```

```python
# –∑–º—ñ–Ω—é—î–º–æ –µ–ª–µ–º–µ–Ω—Ç–∏ –ø–æ —ñ–Ω–¥–µ–∫—Å–∞–º –Ω–∞ nan
df[repl_df[:,0],repl_df[:,1]] = np.nan
print(df)
```

–ó–Ω–∞–π—Ç–∏ –ø–æ–∑–∏—Ü—ñ—ó –≤—Å—Ç–∞–≤–ª–µ–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω—å np.nan –≤ 1-–π –∫–æ–ª–æ–Ω—Ü—ñ

```python
if True in np.isnan(df[:,0]):
    index = np.where(np.isnan(df[:,0]))
    print('–Ü–Ω–¥–µ–∫—Å–∞ –ø–æ–∑–∏—Ü—ñ–π nan :\n', index[0])
    position = index[0] + 1
    print('–ü–æ–∑–∏—Ü—ñ—è —É–ª–µ–º–µ–Ω—Ç–∞ nan :\n', position)
```

–í—ñ–¥—Ñ—ñ–ª—å—Ç—Ä—É–≤–∞—Ç–∏ –º–∞—Å—Å–∏–≤ –∑–∞ —É–º–æ–≤–æ—é: –∑–Ω–∞—á–µ–Ω–Ω—è –≤ 3-–π –∫–æ–ª–æ–Ω—Ü—ñ > 1.5 —Ç–∞ –∑–Ω–∞—á–µ–Ω–∏—è –≤ 1-–π –∫–æ–ª–æ–Ω—Ü—ñ < 5.0 (–∑–±–µ—Ä–µ–≥—Ç–∏ —É —ñ–Ω—à—É –∑–º—ñ–Ω–Ω—É)

```python
filter_df = df[np.logical_and(df[:,2] > 1.5, df[:,0] < 5.0)]
print(filter_df)
```

–ó–∞–º—ñ–Ω–∏—Ç–∏ –≤—Å—ñ –∑–Ω–∞—á–µ–Ω–Ω—è np.nan –Ω–∞ 0

```python
df[np.isnan(df)] = 0
print(df)
```

–ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ –≤—Å—ñ —É–Ω—ñ–∫–∞–ª—å–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –≤ –º–∞—Å—Å–∏–≤—ñ —Ç–∞ –≤–∏–≤–µ—Å—Ç–∏ —ó—Ö —Ä–∞–∑–æ–º —ñ–∑ –∫—ñ–ª—å–∫—ñ—Å—Ç—é

```python
uniq_value = np.unique(df, return_counts=True)
# –í–∏–≤–æ–¥–∏–º–æ —É–Ω—ñ–∫–∞–ª—å–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è
print('–£–Ω—ñ–∫–∞–ª—å–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è : ', uniq_value[0])
# –í–∏–≤–æ–¥–∏–º–æ —ó—Ö –∫—ñ–ª—å–∫—ñ—Å—Ç—å
print('–ö—ñ–ª—å–∫—ñ—Å—Ç—å –∑–Ω–∞—á–µ–Ω—å : ', uniq_value[1])
```

–†–æ–∑–±–∏—Ç–∏ –º–∞—Å–∏–≤ –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—ñ –Ω–∞ 2 —Ä—ñ–≤–Ω—ñ —á–∞—Å—Ç–∏–Ω–∏ (–Ω–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –∞–±—Å–æ–ª—é—Ç–Ω—ñ —á–∏—Å–ª–∞, –º–∞—é—Ç—å –±—É—Ç–∏ –¥–≤–∞ –º–∞—Å—Å–∏–≤–∏ –ø–æ 4 –∫–æ–ª–æ–Ω–∫–∏)

```python
two_df = np.vsplit(df,2)
first_df = two_df[0]
second_df = two_df[1]
#–ø–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ —â–æ –¥—ñ–π—Å–Ω–æ 2 –º–∞—Å–∏–≤–∞ –ø–æ 75 —Å—Ç—Ä–æ–∫ —Ç–∞ 2 –∫–æ–ª–æ–Ω–∫–∏
print('–†–æ–∑–º—ñ—Ä –ø–µ—Ä—à–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ : ', first_df.shape)
print('–†–æ–∑–º—ñ—Ä –¥—Ä—É–≥–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ : ', second_df.shape)
```

–í—ñ–¥—Å–æ—Ä—Ç—É–≤–∞—Ç–∏ –æ–±–∏–¥–≤–∞ –º–∞—Å—Å–∏–≤–∏ –ø–æ 1-–π –∫–æ–ª–æ–Ω—Ü—ñ: 1-–π –∑–∞ –∑–±—ñ–ª—å—à–µ–Ω–Ω—è–º, 2-–π –∑–∞ –∑–º–µ–Ω—à–µ–Ω–Ω—è–º

```python
# —Å–æ—Ä—Ç—É—î–º–æ –ø–µ—Ä—à–∏–π –º–∞—Å—Å–∏–≤ –∑–∞ –∑–±—ñ–ª—å—à–µ–Ω–Ω—è–º
min_max = first_df[np.argsort(first_df[:,0])]
# —Å–æ—Ä—Ç—É—î–º–æ –¥—Ä—É–≥–∏–π –º–∞—Å–∏–≤ –∑–∞ –∑–º–µ–Ω—à–µ–Ω–Ω—è–º (—Å–æ—Ä—Ç—É—î–º–æ —Ç–∞ –∑–¥—ñ–π—Å–Ω—é—î–º–æ —Ä–µ–≤–µ—Ä—Å)
max_min = second_df[np.argsort(second_df[:,0])][::-1,]

print('–í—ñ–¥—Ñ—ñ–ª—å—Ç—Ä–æ–≤–∞–Ω–∏–π –º–∞—Å–∏–≤ –∑–∞ –∑–º–µ–Ω—à–µ–Ω–Ω—è–º :\n', min_max, '\n')
print('–í—ñ–¥—Ñ—ñ–ª—å—Ç—Ä–æ–≤–∞–Ω–∏–π –º–∞—Å–∏–≤ –∑–∞ –∑–±—ñ–ª—å—à–µ–Ω–Ω—è–º :\n', max_min)
```

–ó—ñ–±—Ä–∞—Ç–∏ –æ–±–∏–¥–≤–∞ –º–∞—Å—Å–∏–≤–∏ –≤ –æ–¥–Ω–µ —Ü—ñ–ª–µ

```python
aft_sort_df = np.concatenate([min_max, max_min], axis=0)
print("–ú–∞—Å–∏–≤ –ø—ñ—Å–ª—è –æ–±'—î–¥–Ω–∞—è–Ω–Ω—è : \n", aft_sort_df)
```

–ó–Ω–∞–π—Ç–∏ –Ω–∞–π–±—ñ–ª—å—à —á–∞—Å—Ç–æ –ø–æ–≤—Ç–æ—Ä—é–≤–∞–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è –≤ –º–∞—Å—Å–∏–≤—ñ

```python
#–∑–Ω–∞—Ö–æ–¥–∏–º–æ —ñ–Ω–¥–µ–∫—Å –±—ñ–ª—å—à –ø–æ–≤—Ç–æ—Ä—é–≤–∞–ª—å–Ω–æ–≥–æ —á–∏—Å–ª–∞ —ñ–∑ –º–∞—Å–∏–≤–∞ uniq_value
more_repeat_ind = np.where(uniq_value[1] == uniq_value[1].max())[0]
#–∑–Ω–∞—Ö–æ–¥–∏–º–æ –ø–æ —ñ–Ω–¥–µ–∫—Å—É –∑–Ω–∞—á–µ–Ω–Ω—è —Ç–∞ –≤–∏–≤–æ–¥–∏–º–æ –π–æ–≥–æ
more_repeat_val = uniq_value[0][more_repeat_ind]
print(more_repeat_val)
```

–ù–∞–ø–∏—Å–∞—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—é, —è–∫–∞ –± –º–Ω–æ–∂–∏–ª–∞ –≤—Å—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –≤ –∫–æ–ª–æ–Ω—Ü—ñ, —è–∫—ñ –º–µ–Ω—à–µ —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Ü—ñ–π –∫–æ–ª–æ–Ω—Ü—ñ, –Ω–∞ 2, —ñ –¥—ñ–ª–∏–ª–∞ —ñ–Ω—à—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –Ω–∞ 4.

```python
# –∑–≤–∏—á–∞–π–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—ñ—è –¥–ª—è –∑–∞—Å—Ç–æ—Å—É–≤–∞–Ω–Ω—è vectorize
def do_matematic(mn, masiv):
    if masiv < mn:
        masiv = masiv * 2
    else:
        masiv = masiv / 4
    return masiv
```

–ó–∞—Å—Ç–æ—Å—É–≤–∞—Ç–∏ –æ—Ç—Ä–∏–º–∞–Ω—É —Ñ—É–Ω–∫—Ü—ñ—é –¥–æ 3-—ó –∫–æ–ª–æ–Ω–∫–∏

```python
final_column = aft_sort_df[:,2]
mn_final_column = np.mean(final_column)
vect_do_matematic = np.vectorize(do_matematic, otypes=[float])
final_columns = vect_do_matematic(mn_final_column, final_column)
print('–ü—Ä–∏–∫–ª–∞–¥ —Ä–æ–±–æ—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó –∑ vectorize :\n',final_columns)
```

–ê–Ω–∞–ª–æ–≥—ñ—á–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –±–µ–∑ vectorize

```python
def mathematic(col):
    mn = np.mean(col)
    new_col = np.where(col < mn, col*2, col/4)
    return new_col
```

```python
non_vectorize = mathematic(aft_sort_df[:,2])
print(non_vectorize)
```

```python
#–ø–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ —á–∏ –¥–æ—Ä—ñ–≤–Ω—é—é—Ç—å –æ–¥–Ω–∞ –æ–¥–Ω—ñ–π:
if final_columns.all() == non_vectorize.all():
    print('Congratulation ü•≥')
```
